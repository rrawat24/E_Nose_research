```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
import numpy as np
import os 
os.getcwd()
```

```python

# Load the data
data = pd.read_excel("DATALOG with Cleaning.xlsx", sheet_name="Combined_gases")
data.head()
```

```python


plt.hist(data['Ammonia (NH3)-Echem'], bins=20, alpha=0.5, label='Ammonia (NH3)-Echem')
plt.hist(data['Nitrogen Dioxide (NO2)'], bins=20, alpha=0.5, label='Nitrogen Dioxide (NO2)')
plt.hist(data['Ethylene'], bins=20, alpha=0.5, label='Ethylene')
plt.hist(data['Carbon Monoxide (CO)'], bins=20, alpha=0.5, label='Carbon Monoxide (CO)')
plt.hist(data['Alcohol- Acetone'], bins=20, alpha=0.5, label='Alcohol- Acetone')
plt.hist(data['Alcohol'], bins=20, alpha=0.5, label='Alcohol')
plt.hist(data['Ammonia (NH3)- Ethanol'], bins=20, alpha=0.5, label='Ammonia (NH3)- Ethanol')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Gases')
plt.legend(loc='upper right')
plt.show()
```

```python

# Extract the feature columns
feature_cols = ["Ammonia (NH3)-Echem", "Nitrogen Dioxide (NO2)", "Ethylene", "Carbon Monoxide (CO)", "Alcohol- Acetone", "Alcohol", "Ammonia (NH3)- Ethanol"]

# Extract the target column
target_col = "Gas Type(ppm)"

# Create the feature matrix X and target vector y
X = data[feature_cols]
y = data[target_col]

# Create an instance of the LDA classifier
lda = LinearDiscriminantAnalysis()

# Fit the LDA model to the data
lda.fit(X, y)

# Predict the class labels for the feature matrix X
y_pred = lda.predict(X)

# Print the accuracy of the LDA model
accuracy = lda.score(X, y)
print("Accuracy:", accuracy)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_pred)
plt.xlabel(feature_cols[0])
plt.ylabel(feature_cols[1])
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_pred_encoded = le.fit_transform(y_pred)

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_pred_encoded, cmap='viridis')
plt.xlabel(feature_cols[0])
plt.ylabel(feature_cols[1])
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

le = LabelEncoder()
y_pred_encoded = le.fit_transform(y_pred)

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_pred_encoded, cmap='viridis')
plt.xlabel(feature_cols[0])
plt.ylabel(feature_cols[1])

# Get unique labels in y_pred_encoded
unique_labels = le.inverse_transform(list(set(y_pred_encoded)))

# Create a list of scatter plot handles and labels for each class
handles = []
labels = []
for i, label in enumerate(unique_labels):
    handles.append(plt.scatter([],[], c='viridis', label=label))
    labels.append(label)

# Add legend
plt.legend(handles, labels, loc='best')

plt.show()
```
